```python
def get_header_data(text):
    header_pattern = r"\s+(?P<service_name>.+\d*)\s+\((?P<pid>\d+)\),\s+(?P<events>\d+)\s+events,\s+(?P<cpu_usage>\d+\.\d+)%"
    match = re.match(header_pattern, text)
    if match:
        return match.groupdict()
```

```python
def get_sub_process(text):
    pattern = "\s*(?P<syscall>\S+)\s+(?P<calls>\d+)\s+(?P<total>\d+(\.\d+)?)\s+(?P<min>\d+(\.\d+)?)\s+(?P<avg>\d+(\.\d+)?)\s+(?P<max>\d+(\.\d+)?)\s+\s+(?P<stddev>\d+(\.\d+)?)%"
    matches = re.finditer(pattern, text)

    return [match.groupdict() for match in matches]
```

```python
import time
import numpy as np

def is_timestamp_in_microseconds(df: pd.DataFrame) -> bool:
    max_timestamp = df["timestamp"].max()
    return max_timestamp > time.time()

df["timestamp"]= df["timestamp"].astype(float)

if "timestamp" in df.columns:
    pass
elif "time" in df.columns:
    df["timestamp"] = df["time"]
elif "Time" in df.columns:
    df["timestamp"] = df["Time"]

if is_timestamp_in_microseconds(df):
    # df['timestamp'] = pd.to_datetime(df['timestamp'], unit='ms')
    df["timestamp"] = pd.to_datetime(np.floor(df["timestamp"] / 1000), unit="s")
else:
    df["timestamp"] = pd.to_datetime(df["timestamp"], unit="s")

df.set_index("timestamp", inplace=True, drop=True)

df.drop(columns=["time", "Time", "timestamp"], inplace=True, errors="ignore")
df.sort_index(inplace=True)
```

```python
df
```

```python
from collections import defaultdict

series = []
for timestamp, group_df in df.resample("20S"):
    processes_store = defaultdict(dict)
    for index, process_array in group_df.iterrows():
        print(index)
        
        for process in process_array["processes"]:
            pid = process["pid"]
            processes_store[pid]["pid"] = pid
            processes_store[pid]["max_cpu_usage"] = max(processes_store[pid].get("cpu_usage",0), float(process["cpu_usage"]))
            processes_store[pid]["cpu_usage"] = processes_store[pid].get("cpu_usage",0) + float(process["cpu_usage"])
            processes_store[pid]["occurances"] = processes_store[pid].get("occurances",0) + 1
            processes_store[pid]["events"] = processes_store[pid].get("events",0) + int(process["events"])
            processes_store[pid]["system_calls"] = processes_store[pid].get("system_calls","") +" "+" ".join([call["syscall"] for call in process["system_calls"]])
            processes_store[pid]["service_name"] = process["service_name"]
    
    features = {"timestamp": timestamp, "processes": list(processes_store.values())}
    series.append(features)
```

```python
df_mean = pd.DataFrame(series).set_index("timestamp", inplace=False, drop=True)
```

```python
df_mean.to_csv("mean_data.csv")
```
